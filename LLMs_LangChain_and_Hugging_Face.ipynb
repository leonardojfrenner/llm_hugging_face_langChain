```python
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline, BitsAndBytesConfig
```

```python
import torch
import getpass
import os

device = "cuda:0" if torch.cuda.is_available() else "cpu"
```

```python
device
```

```python
torch.random.manual_seed(42)
```

```python
os.environ["HF_TOKEN"] = getpass.getpass()
```

```python
id_model = "microsoft/Phi-3-mini-4k-instruct"
```

```python
model = AutoModelForCausalLM.from_pretrained(
    id_model,
    device_map = "cuda",
    torch_dtype = "auto",
    trust_remote_code = True,
    attn_implementation = "eager"
)
```

```python
tokenizer = AutoTokenizer.from_pretrained(id_model)
```

```python
pipe = pipeline("text-generation",model = model, tokenizer = tokenizer)
```

```python
generation_args = {
    "max_new_tokens": 500,
    "return_full_text":False,
    "temperature":0.1,
    "do_sample": True,
}
```

```python
prompt = "Explique o que é computação quântica"
output = pipe(prompt, **generation_args)
```

```python
def generate_text(pipe, **generation_args):
    prompt = input("Digite seu prompt: ")
    template = """<|system|> You are a helpful assistant.<|end|> <|user|> "{}"<|end|> <|assistant|>""".format(prompt)
    output = pipe(template, **generation_args)
    return output[0]['generated_text']
```

```python
generate_text(pipe, **generation_args)
```

```python
#com um modelo maior usando tecnicas de quantização
```

```python
quantization_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_use_double_quant=True,
    bnb_4bit_compute_dtype=torch.bfloat16
)
```

```python
model_id = "meta-llama/Meta-Llama-3-8B-Instruct"

model = AutoModelForCausalLM.from_pretrained(model_id, quantization_config=quantization_config)
tokenizer = AutoTokenizer.from_pretrained(model_id)
```

```python
prompt = ("quem foi a primeira pessoa no espaço?")
messages = [{"role": "user", "content": prompt}]
```

```python
encodeds = tokenizer.apply_chat_template(messages, return_tensors="pt")
model_inputs = encodeds.to(device)
generated_ids = model.generate(model_inputs, max_new_tokens=1000, do_sample=True, pad_token_id=tokenizer.eos_token_id)

decoded = tokenizer.batch_decode(generated_ids)
res = decoded[0]
res
```
